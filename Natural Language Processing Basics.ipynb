Given the text file (grail.txt) from the Web Text Corpus of NLTK, perform the following tasks:
1. Report the number of sentences and tokens contained in the file given as input.
2. Convert the whole text to lower case and report the number of unique tokens present
before and after lower casing in the input file.
3. Report the number of stopwords in the file. Report the number of tokens left after
stopword removal.
4. Perform stemming after removing stopwords and report the number of unique tokens left
in the text.
5. Report the number of words starting with a consonant and the number of words starting
with a vowel in the file given after performing steps 1,2,3, and 4.
6. Given a word and a file as input, return the number of sentences starting with that word
in the input file after performing steps 1,2,3, and 4.

7. Given a word and a file as input, return the number of sentences ending with that word in
the input file after performing steps 1,2,3, and 4.
8. Given a word and a file as input, return the count of that word in the input file after
performing steps 1,2,3, and 4.

```python
#Importing necessary libraries

import nltk
import string
import re
from nltk.stem import PorterStemmer
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize, sent_tokenize
from collections import Counter
nltk.download('stopwords')
nltk.download('punkt')
```

Reading a text file named grail.txt and printing the same.

```python
# reading file grail.txt and printing it.

with open("grail.txt") as f:
  t = f.read()
print(t)
```

### 1. Report the number of sentences and tokens contained in the file given as input.

```python
#to get the number of sentences len() will give the count

sentences = sent_tokenize(t)

#to get number of tokens, applying len() will the count as follows

token = word_tokenize(t)


print("The number of sentences in the file is", len(sentences))
print("The number of tokens in the file is", len(token))
```

### 2. Convert the whole text to lower case and report the number of unique tokens present before and after lower casing in the input file.

```python
#Unique tokens before lower case

utokens = list(set(token))
utokens.sort()
#print(utokens)
print('The number of unique tokens before lower casing is', len(utokens))
```

```python
#Applying lower casing 

t_lower = t.lower()
print(t_lower)  # To check if the text got lower cased.
```

```python
#Unique tokens after lower case

lower_token = word_tokenize(t_lower)
utokens_after_lower = list(set(lower_token))
utokens_after_lower.sort()
#print(utokens_after_lower)
print('The number of unique tokens after lower casing is', len(utokens_after_lower))
```

### 3. Report the number of stopwords in the file. Report the number of tokens left after stopword removal.

```python
#Gathering stopwords 

stop_words = set(stopwords.words('english'))

# Keeping non stopwords in one variable
without_stop_words = [x for x in utokens_after_lower if x not in stop_words]

# Keeping stopwords in one variable
with_stop_words = [x for x in utokens_after_lower if x in stop_words]


# 113 + 1726 = 1839 So we have segragated the stopwords & rest of the token in our file correctly
print("The number of stopwords in our text file is %d & the number of tokens left after stopword removal is %d" %(len(with_stop_words), len(without_stop_words)))
```

### 4. Perform stemming after removing stopwords and report the number of unique tokens left in the text.

```python
# Performing stemming on the list containing no stopwords.

Porter_Stemmer = PorterStemmer()
Stemmed = [Porter_Stemmer.stem(a) for a in without_stop_words]

Uniques = list(set(Stemmed))

Uniques.sort()   # Not necessary. Just for a better view if printing the next line.
#print(Uniques)  # uncomment to check the unique tokens left after stemming

print('The number of unique tokens after stemming is %d' %(len(Uniques)))
```

### 5. Report the number of words starting with a consonant and the number of words starting with a vowel in the file given after performing steps 1,2,3, and 4.

```python
# using regular expression to match consonant & vowels & counting using len()

print("The number of words starting with a consonant is %d and \nthe number of words starting with a vowel in our file after performing steps 1, 2, 3 and 4 is %d" %(len([x for x in Uniques if(re.search('^[bcdfghjklmnpqrstvwxyzBCDFGHJKLMNPQRSTVWXYZ][A-Za-z0-9_]*', x))]), len([x for x in Uniques if(re.search('^[aeiouAEIOU][A-Za-z0-9_]*', x))])))
```
